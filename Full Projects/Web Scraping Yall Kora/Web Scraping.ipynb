```python
# load work page
import requests
# extract data from html pages
from bs4 import BeautifulSoup
import csv
```

```python
# Month = input("Please Enter A Month: ")
pages = []
dates = []
for i in range(1,46):
    if i <=31:
        Month = 1
    else:
        Month = 2
        i+=1
    pages.append(requests.get(f"https://www.yallakora.com/match-center/?date={Month}/{i%32}/2024"))
    date = f"{Month}/{i%32}/2024"
    dates.append(date)
```

```python
Info = {"champion_title":[],"first_team":[],"second_team":[],"hour":[],"score_1":[],"score_2":[],"Date":[]}
```

```python
Info = {"Date":[],"champion_title":[],"first_team":[],"second_team":[],"hour":[],"score_1":[],"score_2":[]}
def main(pages):
    for a,n in enumerate(pages):
        src = n.content
        soup = BeautifulSoup(src,"lxml")
        champions = soup.find_all("div",{"class":"matchCard"})
        NumChampions = len(champions)

    #     print(NumChampions)
        def get_match_info(champions):
            for i in range(NumChampions):
                all_matches = champions[i].find_all("div",{"class":"liItem"})
                numMatches = len(all_matches)
                for j in range(numMatches):
                    Info["Date"].append(dates[a])
                    champion_title = champions[i].find("h2").text.strip()
                    Info["champion_title"].append(champion_title)
                    first_team = all_matches[j].find("div",{"class":"allData"}).find("div",{"class":"teamsData"}).find("div",{"class":"teamA"}).find("p").text.strip()
                    Info["first_team"].append(first_team)
                    second_team = all_matches[j].find("div",{"class":"allData"}).find("div",{"class":"teamsData"}).find("div",{"class":"teamB"}).find("p").text.strip()        
                    Info["second_team"].append(second_team)
                    hour = all_matches[j].find("div",{"class":"allData"}).find("div",{"class":"teamsData"}).find("div",{"class":"MResult"}).find("span",{"class":"time"}).text.strip()
                    Info["hour"].append(hour)
                    score_1 = all_matches[j].find("div",{"class":"allData"}).find("div",{"class":"teamsData"}).find("div",{"class":"MResult"}).find_all("span",{"class":"score"})[0].text.strip()
                    Info["score_1"].append(score_1)
                    score_2 = all_matches[j].find("div",{"class":"allData"}).find("div",{"class":"teamsData"}).find("div",{"class":"MResult"}).find_all("span",{"class":"score"})[1].text.strip()
                    Info["score_2"].append(score_2)


        get_match_info(champions)
```

```python
main(pages)
```

```python
import pandas as pd
df = pd.DataFrame(Info)
```
