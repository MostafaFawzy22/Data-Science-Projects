```python
1. العمر: عمر الموظف.
2. الاستنزاف: يشير إلى ما إذا كان الموظف قد ترك الشركة أو لا يزال يعمل.
3. BusinessTravel: يصف تكرار رحلات العمل للموظف (على سبيل المثال، "السفر نادرًا" أو "السفر بشكل متكرر" أو "بدون سفر").
4. DailyRate: معدل الأجر اليومي للموظف.
5. القسم: القسم الذي يعمل فيه الموظف.
6. المسافة من المنزل: المسافة بالأميال بين منزل الموظف ومكان عمله.
7. التعليم: أعلى مستوى تعليمي حصل عليه الموظف (يتراوح من 1 إلى 5 ويمثل المستويات التعليمية المختلفة).
8. المجال التعليمي: مجال التعليم الذي درسه الموظف.
9. عدد الموظفين: عدد الموظفين بنفس العدد (عادةً 1 لجميع الصفوف).
10. رقم الموظف: معرف فريد مخصص لكل موظف.
11. الرضا البيئي: مستوى رضا الموظف عن بيئة عمله (يتراوح من 1 إلى 4، وهو ما يمثل مستويات مختلفة من الرضا).
12. الجنس: جنس الموظف.
13. معدل الأجر بالساعة: معدل الأجر بالساعة للموظف.
14. المشاركة في العمل: مستوى مشاركة الموظف في العمل (يتراوح من 1 إلى 4، وهو ما يمثل مستويات مختلفة من المشاركة).
15. JobLevel: المستوى الوظيفي للموظف داخل الشركة (يتراوح من 1 إلى 5، ويمثل المستويات الوظيفية المختلفة).
16. JobRole: الدور أو المنصب المحدد للموظف داخل الشركة.
17. الرضا الوظيفي: مستوى الرضا الوظيفي لدى الموظف (يتراوح من 1 إلى 4، وهو ما يمثل مستويات مختلفة من الرضا).
18. الحالة الاجتماعية: الحالة الاجتماعية للموظف.
19. الدخل الشهري: الدخل الشهري للموظف.
20. السعر الشهري: معدل الأجر الشهري للموظف.
21. NumCompaniesWorked: عدد الشركات التي عمل فيها الموظف قبل الوظيفة الحالية.
22. Over18: يشير إلى ما إذا كان عمر الموظف يزيد عن 18 عامًا (عادةً ما يكون "Y" لجميع الصفوف).
23. العمل الإضافي: يشير إلى ما إذا كان الموظف يعمل لوقت إضافي أم لا (عادةً "نعم" أو "لا").
24. PercentSalaryHike: نسبة الزيادة في راتب الموظف مقارنة بالعام السابق.
25. تقييم الأداء: تقييم أداء الموظف (يتراوح من 1 إلى 4، ويمثل مستويات مختلفة من الأداء).
26. الرضا عن العلاقات: مستوى رضا الموظف عن علاقاته في العمل (يتراوح من 1 إلى 4، وهو ما يمثل مستويات مختلفة من الرضا).
27. StandardHours: العدد القياسي لساعات العمل للموظف (عادة 80 لجميع الصفوف).
28. StockOptionLevel: مستوى خيارات الأسهم الممنوحة للموظف (يتراوح من 0 إلى 3، مما يشير إلى مستويات مختلفة من خيارات الأسهم).
29. إجمالي سنوات العمل: إجمالي عدد السنوات التي عمل فيها الموظف.
30. TrainingTimesLastYear: عدد المرات التي تلقى فيها الموظف التدريب في العام السابق.
31. التوازن بين العمل والحياة: التوازن المتصور بين العمل والحياة الخاصة بالموظف (يتراوح من 1 إلى 4، وهو ما يمثل مستويات مختلفة من التوازن).
32. عدد السنوات في الشركة: عدد السنوات التي قضاها الموظف في الشركة.
33. YearsInCurrentRole: عدد السنوات التي قضاها الموظف في منصبه الحالي.
34. YearsSinceLastPromotion: عدد السنوات منذ آخر ترقية للموظف.
35. YearsWithCurrManager: عدد السنوات التي قضاها الموظف في العمل مع مديره الحالي.
```
