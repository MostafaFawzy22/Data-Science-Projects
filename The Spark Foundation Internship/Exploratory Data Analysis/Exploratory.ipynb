```python
## preprocessing
```

```python
q1 = df['Sales'].quantile(0.25)
q3 = df['Sales'].quantile(0.75)
iqr = q3 - q1
min_outliers_sales = q1 - 1.5*iqr
max_outliers_sales = q3 + 1.5*iqr
print(max_outliers_sales,min_outliers_sales)
```

```python
q1 = df['Profit'].quantile(0.25)
q3 = df['Profit'].quantile(0.75)
iqr = q3 - q1
min_outliers = q1 - 1.5*iqr
max_outliers = q3 + 1.5*iqr
print(max_outliers,min_outliers)
```

```python
df_without_outliers = df[(df['Sales']<max_outliers_sales) & (df['Sales']> min_outliers_sales) & (df['Profit']<max_outliers) & (df['Profit']> min_outliers)]
```

1. Ship Mode: This column indicates the mode or method of shipment for the products, such as standard shipping, express shipping, etc.

2. Segment: This column categorizes the customers into different segments based on certain criteria, such as consumer, corporate, or home office.

3. Country: This column represents the name of the country where the sale occurred or the customer is located.

4. City: This column contains the name of the city where the sale occurred or the customer is located.

5. State: This column represents the name of the state or province where the sale occurred or the customer is located.

6. Postal Code: This column contains the postal code or ZIP code associated with the city or address of the sale or customer.

7. Region: This column indicates the geographical region or area of the country where the sale occurred or the customer is located.

8. Category: This column categorizes the products into different categories, such as office supplies, technology, or furniture.

9. Sub-Category: This column provides further classification or subcategories within the main product categories, such as chairs, tables, or phones.

10. Sales: This column represents the total sales revenue generated from the specific transaction or sale.

11. Quantity: This column indicates the number of units or quantity of products sold in the specific transaction.

12. Discount: This column represents the discount applied to the transaction or sale, usually expressed as a percentage.

13. Profit: This column indicates the profit earned from the specific transaction or sale, which is calculated by subtracting the cost from the sales revenue.
