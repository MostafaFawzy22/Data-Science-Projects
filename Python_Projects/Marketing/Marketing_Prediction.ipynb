## Preprocessing for Analysing

```python
# Age Of Customer 2014 - Year_Birth
df["Age"] = 2015 - df["Year_Birth"]
# Study Level Of Customer Graduate & Ungraduate & Postgraduate
df["Level_Education"] = df["Education"].map(
    {
        "Graduation": "Graduate",
        "Master": "Postgraduate",
        "PhD": "Postgraduate",
        "Basic": "Ungraduate",
        "2n Cycle": "Ungraduate",
    }
)
# Status of Customer Married:2 & Single:1
df["Status"] = df["Marital_Status"].map(
    {
        "Married": "Maried",
        "Together": "Maried",
        "Divorced": "Single",
        "Single": "Single",
        "Widow": "Single",
        "Alone": "Single",
        "Absurd": "Single",
        "YOLO": "Single",
    }
)
# Size Of Family = Parents + Children
df["Size_Family"] = (
    df["Status"].map({"Maried": 2, "Single": 1}) + df["Kidhome"] + df["Teenhome"]
)
# Num Of Children
df["Children"] = df["Kidhome"] + df["Teenhome"]
# IS Customer Is Parent or Not
df["Is_Parent"] = np.where(df.Children > 0, 1, 0)
## Period Of Customer in Company
df["Dt_Customer"] = pd.to_datetime(df["Dt_Customer"], format="mixed")
df["Customer_For"] = df["Dt_Customer"].max() - df["Dt_Customer"]
# Amount Of Spending Customer On Products
df["Spent"] = (
    df["MntWines"]
    + df["MntFruits"]
    + df["MntMeatProducts"]
    + df["MntFishProducts"]
    + df["MntSweetProducts"]
    + df["MntGoldProds"]
)
# a feature that combines acceptedcmp 1 - 5. This feature is made to see customer intensity in accepting campaigns
df["total_accepted_campaign"] = (
    df["AcceptedCmp1"]
    + df["AcceptedCmp2"]
    + df["AcceptedCmp3"]
    + df["AcceptedCmp4"]
    + df["AcceptedCmp5"]
)
# a feature created by adding up the total customer acceptance in all previous campaigns (1-5) with the total revenue per accepted campaign
df["total_revenue"] = df["total_accepted_campaign"] * df["Z_Revenue"]
# total purchase
df["total_order"] = (
    df["NumDealsPurchases"]
    + df["NumWebPurchases"]
    + df["NumCatalogPurchases"]
    + df["NumStorePurchases"]
)
# Age_category customer menurut WHO

cut_points_age = [0, 35, 55, 100]
labels = ["Young", "Middle-Age", "Elderly"]
# Create a new column 'Income Category' based on the income column
df["Age Category"] = pd.cut(
    df["Age"], bins=cut_points_age, labels=labels, include_lowest=True
)
# Income
cut_points_income = [0, 20000, 40000, float("inf")]
labels = ["Low Income", "Average Income", "High Income"]
# Create a new column 'Income Category' based on the income column
df["Income Category"] = pd.cut(
    df["Income"], bins=cut_points_income, labels=labels, include_lowest=True
)
```

## Outliers

```python
q3_Spent = df["Spent"].describe()[6]
q1_Spent = df["Spent"].describe()[4]
iqr_Spent = q3_Spent - q1_Spent
outliers_Spent = q3_Spent + 1.5 * iqr_Spent
outliers_Spent
q3_income = df["Income"].describe()[6]
q1_income = df["Income"].describe()[4]
iqr_income = q3_income - q1_income
outliers_income = q3_income + 1.5 * iqr_income
```

```python
df = df[
    (df["Age"] < 90) & (df["Spent"] < outliers_Spent) & (df["Income"] < outliers_income)
]
```
